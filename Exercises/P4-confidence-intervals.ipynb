### Exercise 1
The heat capacity for a liquid is measured as follows:

|kcal/gC°|.645|.654|.640|.627|.626|.649|.629|.631|.643|.633|.646|.630|.634|.631|.651|.659|.638|.645|.655|.624|.658|.658|.658|.647|.665|
|--------|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|

Past experience indicates that $\sigma=.01$.

a) Evaluate $\bar{X}$, thereby obtaining an unbiased point estimate for $\mu$.


```python
import numpy as np

data = [
    0.645, 0.654, 0.640, 0.627, 0.626, 0.649, 0.629, 0.631,
    0.643, 0.633, 0.646, 0.630, 0.634, 0.631, 0.651, 0.659,
    0.638, 0.645, 0.655, 0.624, 0.658, 0.658, 0.658, 0.647, 0.665
]

x_mean = np.mean(data)
x_mean
```


b) Assume that $X$ is normally distributed. Find a 95% confidence interval for $\mu$.


```python
from scipy.stats import norm

n = len(data)
sigma = 0.01
alpha = 0.05
z_95 = norm.ppf(1-alpha/2)

margin = z_95 * sigma / np.sqrt(n)
lower = x_mean - margin
upper = x_mean + margin

lower, upper, z_95
```


c) Would you expect a 90% confidence interval for $\mu$ based on these data to be longer or shorter than the interval in b)?


Jag tror att ett 90 % konfidensintervall kommer ge kortare intervall då z_90 < z_95.

```python
alpha_1 = 0.1
z_90 = norm.ppf(1-alpha_1/2)

z_90
```


d) Would you expect a 99% confidence interval for $\mu$ based on these data to be longer or shorter than the interval in b)? Verify by finding a 99% confindence interval on $\mu$.


Jag tror att ett 99 % konfidensintervall kommer ge längre intervall då z_99 > z_95.

### Exercise 3

Sulfur dioxide causes acid rain. These data were obtained on the concentration in a forest area thought to have been damaged by acid rain:
|µg/m³|52.7|43.9|41.7|71.5|47.6|55.1|62.2|56.5|33.4|61.8|54.3|50.0|45.3|63.4|53.9|65.5|66.6|70.0|52.4|38.6|46.1|44.4|60.7|56.4|
|-----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|----|

a) Compute $\bar{x}$, $s^2$ and $s$, the sample mean, sample variance and sample deviation.

```python
sulfur_data = [
    52.7,   43.9,	41.7,	71.5,	47.6,	55.1,	
    62.2,	56.5,	33.4,	61.8,	54.3,	50.0,
    45.3,	63.4,	53.9,	65.5,	66.6,	70.0,
    52.4,	38.6,	46.1,	44.4,	60.7,	56.4,
]

sulfur_mean = np.mean(sulfur_data)
s2 = np.var(sulfur_data, ddof=1)
s = np.sqrt(s2)

sulfur_mean, s2, s
```


b) Compute a 95% confidence interval on the mean.

```python
from scipy.stats import t

sulfur_n = len(sulfur_data)
t_förd = t.ppf(0.975, df=sulfur_n-1)

sulfur_margin = t_förd * s / np.sqrt(sulfur_n)
sulfur_lower = sulfur_mean - sulfur_margin
sulfur_upper = sulfur_mean + sulfur_margin

sulfur_lower, sulfur_upper
```


c) Given a national average of 20 µg/m³, is there evidence for elevated sulfur levels in the data?

Ja, om det nationella snittet är 20 µg/m³ så ligger detta värde utanför 95 %-konfidensintervallet för det sanna medelvärdet μ.
